```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Link to get the dataset
url= "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv"
```

```python
#load the dataset using pandas dataframe
df = pd.read_csv(url)
df.sample(10)
```

```python
#To see the statitical infromation about the dataset
df.describe()
```

```python
#Select some of the features to be use 
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.sample(20)
```

```python
#Visualizint the histogram of each feature 
viz = cdf[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
viz.hist()
plt.show()
```

```python
#Scatter plotting how eact selected feature correlate to the target variable
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color = 'blue')
plt.xlabel("ENGINESIZE")
plt.ylabel("CO2 EMISSION")
plt.xlim(0, 27)
plt.show()
```

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS, color = 'blue')
plt.xlabel("FUEL CONSUMPTIOPN")
plt.ylabel("CO2 EMISSIONS")
```

```python
#Converting the training set into numpy for the model
x = cdf.ENGINESIZE.to_numpy()
y = cdf.CO2EMISSIONS.to_numpy()
```

```python
#Building simple linear model
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
type(x_train),np.shape(x_train)
```

```python
#Importing the model 
from sklearn import linear_model
regressor = linear_model.LinearRegression()
```

```python
#Fitting the model to get the coefficient and the intercept
regressor.fit(x_train.reshape(-1,1), y_train)
coef_ = regressor.coef_[0]
intercept_ = regressor.intercept_
print("The coefficient is :", coef_)
print("The intercep is : ", intercept_)
```

```python
#Visualizing the model output
plt.scatter(x_train, y_train, color= 'blue')
plt.plot(x_train, coef_ * x_train + intercept_, '-r')
plt.xlabel("Engine Size")
plt.ylabel("CO2 Emissions")
```

```python
#Making the prediction
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(x_train.reshape(-1,1), y_train)
y_pred = regr.predict(x_test.reshape(-1,1))
```

```python
#Using the prediction methodes to predict the test dataset
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("The mean absolute error is: %.2f" % mean_absolute_error(y_test, y_pred))
```

```python


print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
```
